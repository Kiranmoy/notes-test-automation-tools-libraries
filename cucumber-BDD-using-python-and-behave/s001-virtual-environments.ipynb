# What and Why virtual environments?

* A way to keep project dependencies **isolated**.
* Different **versions** of **libraries/packages**.
* Different **versions** of **Python**.
* Multiple projects that depend on the **same library but different versions**.
* Project organizations.
* Keep default Python clean.

# Creating Virtual Environment using `virtualenv`


**"`virtualenv`"** is a package that can be installed with pip
```
pip install virtualenv
```

You can use python 2 pip or python 3 pip (it doesn't matter).

---

**To create virtual environment**:

```
virtualenv <path-to-the-folder-for-virtual-environment>
```

**Or**:

```
virtualenv --python=path-to-oython <path-to-the-folder-for-virtual-environment>
```

---

**To activate the virtual environment**:
* Mac: `source <path-to-virtual-environment-folder>/bin/activate`
* Windows: `<path-to-virtual-environment-folder>/Scripts/activate.bat`

---

**To exit or deactivate the virtual environment, run**: `deactivate`


# How to use virtual environment python without activating?

You can specify the full path of the Python interpreter:
* Mac: **`<apth-to-the-virtual-env-folder>/bin/python`** `my_script.py`
* Windows: **`<apth-to-the-virtual-env-folder>/Scripts/python`** `my_script.py`

This is very useful while running cron jobs or jenkins job.
