**Formatters** is a **behave** module that produces test execution results in various formats.

**`behave --format help`** OR **`behave -f help`**: shows all available formats in which behave can produce output.



# How to output Junit compatible xml?

**`behave --junit --junit-directory junit_output`**

> * For each feature file, one XML will be geenrated
> * Parse the junit xml to generate a CUSTOM REPORT.



# How to output JSON report?

**`behave --format json.pretty -o json_output/my_json_out.json`**

> * One JSON file will be generated for the entire execution.
> * Parse the JSON to generate a CUSTOM REPORT.





# Custom Report Generation

## `report_generator.py`

**Refer the script is present in the current directory for checkout.**



## Execute behave features

**`behave --format json.pretty -o json_output/my_json_out.json`**

## Invoke Custom Report Generation Script

**Directory Structure**



**Custom Report Script arguments**



1. **`--input_json_file`**: Path of input json file. JSON file is output of Behave test run
2. **`--output_html_file`**: Path of the output html file to be generated

**Shell command**

```
python report_generation\custom_report_generator\report_generator.py \
--input_json_file report_generation\json_output\my_json_out.json \
--output_html_file report_generation\custom_report_generator\custom-report.html

```

**Windows command**

```
python report_generation\custom_report_generator\report_generator.py ^
--input_json_file report_generation\json_output\my_json_out.json ^
--output_html_file report_generation\custom_report_generator\custom-report.html

```
