# Command line Options

| SCENARIO | COMMAND |
|:---|:---:|
| Execute all tests in a folder | **`robot .`** |
| Execute specific test | **`robot test_file.robot`** |
| Execute test cases by test name | **`robot -t testname .`** |
| Execute test cases by tags in any file | **`robot --include tagname .`** |
| Execute test cases by multiple tags | **`robot --include tagname1 and tagname2 .`** | 
|                                     | **`robot --include tagname1 or tagname2 .`**  |
| Exclude the tests by tag | **`robot --exclude tagname .`** |
| Execute tests by suite name | **`robot --suite foldername`** |
| Execute only failed tests | **`robot --rerunfailed output.xml .`** |

> **Note**: We can use regex for specifying a test case, or test file, or test folder name.

# Passing values to test from command line

**testDemo.robot**:



**Command to pass `browser_name` value at runtime via command-line**
```
robot --variable browser_name:Firefox testDemo.robot
```

**Note that:**
* The Variable must be declared in the **Variables** section of the robot file.
* The variable value passed in the command line will overwrite the robot file variable declared in the **Variables** section.

# Parallel test execution using `pabot

**Pabot** is a parallel test runner for Robot Framework. It can be used to run tests in parallel on a single machine with multiple processes.

Install Plugin: **`pip install -U robotframework-pabot`**

There are two options for parallel test execution:
1. Execute multiple `robot` files in parallel.
2. Execute multiple test cases in a robot file in parallel.

**Usage**
* By default, Pabot will split execution on the suite level.
* That means that each process will run a single suite.
* Test Cases from the suite will be executed sequentially.
* If you want to split execution on the test level, you can use the `--testlevelsplit` option.
* This will split execution on the test level, so that each process will run a single test case.

**Split execution to suite files:**
```bash
pabot [path to tests]
```

**Split execution on test level:**
```bash
pabot --testlevelsplit [path to tests]
```

**Run with 8 parallel processes:**
```bash
pabot --processes 8 [path to tests]
```

The default number of processes depends on the number of CPU cores. (2 processes per core) If no number pro processes is given, the default number of processes is used.
