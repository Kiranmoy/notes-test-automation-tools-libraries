# Connect Database via Python

````python
import mysql.connector

# host, database, user, password
connection = mysql.connector.connect(host="localhost", database="APIDevelop", user="root", password="root")

print(connection.is_connected())

cursor = connection.cursor()
cursor.execute('select * from CustomerInfo')

row = cursor.fetchone() # first row -> tuple
print("Single ROW: {}".format(row))

all_rows = cursor.fetchall() # list of tuples: second row onwards
print("ALL ROWS")
print(all_rows)

connection.close()
````

# Database Operations

````python
import mysql.connector

# host, database, user, password
connection = mysql.connector.connect(host="localhost", database="APIDevelop", user="root", password="root")

print(connection.is_connected())

cursor = connection.cursor()

# READ OPERATION
cursor.execute('select * from CustomerInfo')
i = 0

# first row -> tuple
row = cursor.fetchone()
print(type(row))
print("Single ROW {}: {}".format(i,row))


# list of tuples: second row onwards
all_rows = cursor.fetchall()
print(type(all_rows))
print("ALL ROWS:")

for row in all_rows:
    i=i + 1
    print("ROW {}: {}".format(i,row))

# UPDATE OPERATION
query = "update customerInfo set Location = %s where CourseName = %s"
data = ("UK", "Jmeter")
cursor.execute(query,data)
connection.commit()

query = "select * from customerInfo where CourseName = %s"
data = ("Jmeter",)
cursor.execute(query,data)
row = cursor.fetchone()
print("ROW UPDATE: {}".format(row))

query = "update customerInfo set Location = %s where CourseName = %s"
data = ("Asia", "Jmeter")
cursor.execute(query,data)
connection.commit()

query = "select * from customerInfo where CourseName = %s"
data = ("Jmeter",)
cursor.execute(query,data)
row = cursor.fetchone()
print("ROW UPDATE: {}".format(row))

connection.close()
````

## configuration.py

````python
import configparser
import mysql.connector
from mysql.connector import Error

def getConfig():
    # read global properties
    config = configparser.ConfigParser()
    config.read("properties.ini")
    return config

def getConnection():
    connect_config = {
        "host": getConfig()["SQL"]["host"],
        "database": getConfig()["SQL"]["database"],
        "user": getConfig()["SQL"]["user"],
        "password": getConfig()["SQL"]["password"]
    }
    # host, database, user, password
    # connection = mysql.connector.connect(host="localhost", database="APIDevelop", user="root", password="root")
    try:
        connection = mysql.connector.connect(**connect_config)
        if(connection.is_connected()):
            print("CONNECTION SUCCESSFUL: {}".format(connect_config))
            return connection
    except Error as err:
        print(err)
````
