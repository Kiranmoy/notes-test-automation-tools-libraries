# Intial setup

```python
from playwright.sync_api import sync_playwright

playwright = sync_playwright().start();

 # Launch a browser (By default, playwright runs in headless=True mode)
browser = playwright.chromium.launch(headless=False,slow_mo=1000)

# Create a new tab/page
page = browser.new_page()

# Goto bootswatch.com
page.goto("https://bootswatch.com/default/")
```

# Role locator

```python
btn = page.get_by_role('button', name='Default button')
btn.highlight()
btn.click()

heading = page.get_by_role('heading', 'Heading 2')
heading.highlight()

radio_btn = page.get_by_role('radio','Option two can be something else and selecting it will deselect option one')
radio_btn.highlight()

checkbox = page.get_by_role('checkbox', 'Default checkbox')
checkbox.highlight()
checkbox.check()
```

# Input Field Locator

```python
email_input = page.get_by_label('Email address')
email_input.highlight()

textarea_input = page.get_by_label('Example textarea')
textarea_input.highlight()

email_input = page.get_by_placeholder('Enter Email')
email_input.highlight()

password_input = page.get_by_placeholder('Enter Password')
password_input.highlight()
```

# Inner Text Locator

* By default, the `get_by_text()` method does partial text matching (i.e., `exact=False`).
* To perform exact text matching use `exact=True` parameter with `get_by_text()` method.

```python
text = page.get_by_text('fine print') # Partial Text Matchin
text = page.highlight()

paragraph = page.get_by_text("fine print", exact=True)  # Exact Text Matching
paragraph.highlight()
```

# Alt Text Locator

```python
page.goto('https://unsplash.com/')

img = page.get_by_alt_text('Point-of-sale system ready to take an order.')
img.highlight()
img.click()
```

# Title Locator

```python
page.goto("https://bootswatch.com/default/")

paragraph = page.get_by_title('attribute')
paragraph.highlight()
```

# CSS Selectors 

CSS Selectors allow to locate elements based on:
* Tagename
* Classname
* Id
* Attribute/Value

> Note that **'css='** is optional because playwright automatically detects whether it's CSS or XPATH locator.

```python
page.locator('css=h1') .highlight()                            # Tagname
page.locator('footer') .highlight()                            # Tagname

page.locator('button.btn-outline-success').highlight()         # Classname
page.locator('button.btn-outline-success').click()             # Classname

page.locator('button#btnGroupDrop1').highlight()               # Id
page.locator('button#btnGroupDrop1').click()                   # Id

page.locator('input[readonly]').highlight()                    # Attribute only
page.locator("input[value='correct value]").highlight()        # Attribute & Value
```

**CSS Selector based on parent-child relationship**

```python
page.locator('nav.bg-dark a.nav-link.active').highlight()
```

**CSS Selector Pseudo classes**

```python
page.locator("h1:text('Nav')").highlight()                        # By default text() function performs partial text matching
page.locator("h1:text-is('Navs')").highlight()                    # To perform exact matching, use test-is() function

page.locator("div.dropdown-menu:visible").highlight()
page.locator("nth-match(button.btn-primary,4)").highlight()
page.locator("nth-match(button:text('primary'),1)").highlight()
```

# XPATH Selector

XPATH SELECTORS allow to locate elements based on:
* Tagename
* Classname
* Id
* Attribute/Value
    
> Absolute XPATH starts with '/'
>
> Relative XPATH starts with '//'
>
> Note that **'xapth='** is optional, playwright automatically detects whether its CSS or XPATH locator.

```python
page.locator("xapth=//h1").highlight()               # Tagname
page.locator("//h1").highlight()                     # Tagname

page.locator("//input[@readonly]").highlight()       # Attribute only
page.locator("//h1[@id='navbars']").highlight()      # Attribute & Value
```

**XPATH Functions**

```python
page.locator("//h1[text()='Heading 1']").highlight()                                   # Exact Text Matching
page.locator("//h1[ contains(text(),'Heading 1') ]").highlight()                       # Partial Text Matching

page.locator("//button[ contains(@class,'btn-outline-primary') ]").highlight()         # Partial Text Matching
page.locator("//input[ contains(@value,'correct') ]").highlight()                      # Partial Text Matching
```

# Locator Chaining

```python
page.get_by_role("button", name="primary").locator("nth=0").highlight()
page.locator("button").locator("nth=18").highlight()

page.get_by_label("Email address").locator("..").highlight()                       # .. selects the parent element

page.locator("div.dropdown-menu").locator("visible=true").highlight()
page.locator("div.dropdown-menu").locator("visible=false").highlight()

# filter() methods check whether the element contains the specified element
page.get_by_role("heading").filter(has_text="Heading").highlight()
page.locator("div.form-group").filter(has=page.get_by_label("Password")).highlight()
```
