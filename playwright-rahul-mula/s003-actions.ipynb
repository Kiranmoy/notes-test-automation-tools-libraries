# Mouse actions

```python

from playwright.sync_api import sync_playwright

with sync_playwright() as playwright:

    browser = playwright.chromium.launch(headless=False,slow_mo=1000)
    page = browser.new_page()
    page.goto("https://bootswatch.com/default/")
    first_block_button = page.get_by_role("button",name='Block Button').first
    last_block_button = page.get_by_role("button", name='Block Button').last
    
    first_block_button.highlight()
    last_block_button.highlight()

    first_block_button.click()
    first_block_button.click(delay=500) # millisecods
    first_block_button.click(button='right') # Right Click
    first_block_button.click(modifiers=['Shift'])
    first_block_button.click(modifiers=["Alt", "Control", "ControlOrMeta", "Meta", "Shift"])

    first_block_button.dblclick()
    first_block_button.dblclick(delay=500) # millisecods
    first_block_button.dblclick(button='right') # Right Click
    first_block_button.dblclick(modifiers=['Shift'])
    first_block_button.dblclick(modifiers=["Alt", "Control", "ControlOrMeta", "Meta", "Shift"])

    first_block_button.hover()

    browser.close()
```

# Fill Inputs

```python
from playwright.sync_api import sync_playwright

with sync_playwright() as plawright:
    browser = plawright.chromium.launch(headless=False, slow_mo=1000)
    page = browser.new_page()
    page.goto("https://bootswatch.com/default/")

    email_input = page.get_by_placeholder("Enter email")

    email_input.fill("me@that.site1")
    print(email_input.input_value()) # Fetch the date from the email input box
    email_input.clear()

    email_input.type("me@that.site2") # Simulate typing (same as fill method)
    print(email_input.input_value())  # Fetch the date from the email input box
    email_input.clear()

    email_input.type("me@that.site3",delay=500) # Simulate typing with delay
    print(email_input.input_value())  # Fetch the date from the email input box
    email_input.clear()


    browser.close()
```

# Radio, Checkboxes & Switch

```python
from playwright.sync_api import sync_playwright

with sync_playwright() as playwright:
    browser = playwright.chromium.launch(headless=False, slow_mo=1000)
    page = browser.new_page()
    page.goto("https://bootswatch.com/default/")

    radio_option_1 = page.get_by_label("Option one is this and that—be sure to include why it's great")
    radio_option_2 = page.get_by_label("ption two can be something else and selecting it will deselect option one")


    radio_option_1.check()
    print("IS Radio Option 1 checked? ", radio_option_1.is_checked())
    print("IS Radio Option 2 checked? ", radio_option_2.is_checked())

    radio_option_2.check()
    print("IS Radio Option 1 checked? ", radio_option_1.is_checked())
    print("IS Radio Option 2 checked? ", radio_option_2.is_checked())

    radio_option_1.click()
    print("IS Radio Option 1 clicked? ", radio_option_1.is_checked())
    print("IS Radio Option 2 clicked? ", radio_option_2.is_checked())

    radio_option_2.click()
    print("IS Radio Option 1 clicked? ", radio_option_1.is_checked())
    print("IS Radio Option 2 clicked? ", radio_option_2.is_checked())

    checkbox_1 = page.get_by_label("Default checkbox")
    checkbox_2 = page.get_by_label("Checked checkbox")

    checkbox_2.click()
    print("IS Checkbox 2 unchecked by click? ", not checkbox_2.is_checked())

    checkbox_1.check()
    print("IS Checkbox 1 checked? ", checkbox_1.is_checked())

    checkbox_1.uncheck()
    print("IS Checkbox 1 checked? ", checkbox_1.is_checked())

    checkbox_1.set_checked(True)
    print("IS Checkbox 1 checked? ", checkbox_1.is_checked())

    checkbox_1.set_checked(False)
    print("IS Checkbox 1 checked? ", checkbox_1.is_checked())

    switch_1 = page.get_by_label("Default switch checkbox input")
    switch_2 = page.get_by_label("Checked switch checkbox input")

    switch_2.click()
    print("IS Switch 2 unchecked by click? ", not switch_2.is_checked())

    switch_1.check()
    print("IS Switch 1 checked? ", switch_1.is_checked())

    switch_1.uncheck()
    print("IS Switch 1 checked? ", switch_1.is_checked())

    switch_1.set_checked(True)
    print("IS Switch 1 checked? ", switch_1.is_checked())

    switch_1.set_checked(False)
    print("IS Switch 1 checked? ", switch_1.is_checked())

    browser.close()
```

# Select option from option menu

```python
from playwright.sync_api import sync_playwright

with sync_playwright() as playwright:
    browser = playwright.chromium.launch(headless=False, slow_mo=1000)
    page = browser.new_page()
    page.goto("https://bootswatch.com/default/")

    select = page.get_by_label("Example select")
    select.select_option("4")
    print("Selected Options are: {}".format(select.input_value()))

    multi_select = page.get_by_label("Example multiple select")
    multi_select.select_option(["1","3","5"])
    print("Selected Options are: {}".format(multi_select.input_value()))

    browser.close()
```

# Dropdown Menu

```python
from playwright.sync_api import sync_playwright

with sync_playwright() as playwright:
    browser = playwright.chromium.launch(headless=False, slow_mo=1000)
    page = browser.new_page()
    page.goto("https://bootswatch.com/default/")

    drowpdown = page.locator("button#btnGroupDrop1")
    drowpdown.click()

    drowpdown_menu = page.locator("div.dropdown-menu:visible")
    drowpdown_menu.highlight()

    # page.locator("div.dropdown-menu:visible a:text('Dropdown link')")
    dropdown_option = drowpdown_menu.locator("a:text('Dropdown link')").last
    dropdown_option.highlight()
    dropdown_option.click()

    browser.close()

```

# Upload files

```python
from playwright.sync_api import sync_playwright

with sync_playwright() as playwright:
    browser = playwright.chromium.launch(headless=False, slow_mo=1000)
    page = browser.new_page()
    page.goto("https://bootswatch.com/default/")

    # SIMPLE FILE UPLOAD INPUT

    file_input = page.get_by_label("Default file input example")
    file_input.set_input_files("file1.txt") # single file path

    # multiple file upload
    # file_input.set_input_files(["file1.txt","file2.txt"])  # multiple file path

    # FILE CHOOSER
    with page.expect_file_chooser() as fc_info:
        file_input.click()
        file_chooser = fc_info.value
        file_chooser.set_files("file2.txt")

    browser.close()
```

# Keyboad events

```python
from playwright.sync_api import sync_playwright

with sync_playwright() as playwright:
    browser = playwright.chromium.launch(headless=False, slow_mo=1000)
    page = browser.new_page()
    page.goto("https://bootswatch.com/default/")

    textarea = page.get_by_label("Example textarea")

    textarea.fill("word")
    print(textarea.input_value())
    textarea.clear()

    textarea.press("Shift+KeyW")
    textarea.press("KeyO")
    textarea.press("KeyR")
    textarea.press("KeyD")

    textarea.press("Control+ArrowLeft") # shortcut to put the cursor at the beginning (left)
    textarea.press("ArrowRight") # move 1 space right

    browser.close()
```
