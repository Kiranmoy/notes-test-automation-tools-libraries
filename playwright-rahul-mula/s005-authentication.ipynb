# Save Authentication State after Google Login

```python
from playwright.sync_api import sync_playwright

with sync_playwright() as playwright:
    browser = playwright.chromium.launch(
        headless=False,
        slow_mo=1000,
        args=["--disable-dev-shm-usage","--disable-blink-features=AutomationControlled"]
        # If browser prevents playwright action and tell "Couldn't Sign You. The browser or app may be unsecure",
        # then use, args=["--disable-dev-shm-usage","--disable-blink-features=AutomationControlled"]
    )

    # Creating context to store the Authentication State and save it on login
    context = browser.new_context()

    page = context.new_page()
    page.goto("https://accounts.google.com")

    # Enter Email
    email_input = page.get_by_label("Email or phone")
    email_input.fill("email@gmail.com")

    page.get_by_role("button", name="Next").click()

    # Enter Password
    password_input = page.get_by_label("Enter your password")
    password_input.fill("email@gmail.com")

    page.get_by_role("button", name="Next").click()

    # Manually complete 2-step verification
    page.pause() # Pause the execution and allow to manually interact with the browser

    # SAVE AUTHENTICATION STATE (AUTHENTICATION COOKIE)

    # Playwright recommends to store it in separate folder structure (i.e. playwright/.auth)
    context.storage_state(
        path="playwright/.auth/google_login_state.json"
    )
    # closing browser context
    context.close()
```

# Reuse the Saved Authentication State for future Login

```python

from playwright.sync_api import sync_playwright

with sync_playwright() as playwright:

    browser = playwright.chromium.launch(
        headless=False,
        slow_mo=1000,
        args=["--disable-dev-shm-usage", "--disable-blink-features=AutomationControlled"]
        # If browser prevents playwright action and tell "Couldn't Sign You. The browser or app may be unsecure",
        # then use, args=["--disable-dev-shm-usage","--disable-blink-features=AutomationControlled"]
    )

    context = browser.new_context(
        storage_state="playwright/.auth/google_login_state.json"
    )

    page = context.new_page()
    page.goto("https://accounts.google.com")

    # Enter Email
    email_input = page.get_by_label("Email or phone")
    email_input.fill("email@gmail.com")

    page.get_by_role("button", name="Next").click()

    # Enter Password
    password_input = page.get_by_label("Enter your password")
    password_input.fill("email@gmail.com")

    page.get_by_role("button", name="Next").click()

    context.close()



```
