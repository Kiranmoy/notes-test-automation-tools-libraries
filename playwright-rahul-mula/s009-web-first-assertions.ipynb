# Page assertions

```python
from playwright.sync_api import Page, expect

DOCS_URL=("https://playwright.dev/python/docs/intro")

def test_page_has_get_started_link(page: Page):
    page.goto("https://playwright.dev/python")

    link = page.get_by_role("link", name="GET STARTED")
    link.click()

    # assert page.url == DOCS_URL
    expect(page).to_have_url(DOCS_URL)
    expect(page).to_have_title("Expected TITLE")
    expect(page).not_to_have_url("Unexpected URL")
    expect(page).not_to_have_title("Unexpected TITLE")

```

# Inner Text assertions

```python
from playwright.sync_api import Page, expect

DOCS_URL=("https://playwright.dev/python/docs/intro")

def test_page_has_get_started_link(page: Page):
    page.goto("https://playwright.dev/python")

    heading = page.locator("h1.hero__title")

    expect(heading).to_contain_text("partial text matching") # Partial text matching: can use regex too
    expect(heading).to_have_text("exact text matching") # Exact text matching

    expect(heading).not_to_contain_text("partial text matching")  # Partial text matching: can use regex too
    expect(heading).not_to_have_text("exact text matching")  # Exact text matching
    
```

# Attribute assertions

```python
import re
from playwright.sync_api import Page, expect

DOCS_URL=("https://playwright.dev/python/docs/intro")

def test_page_has_get_started_link(page: Page):
    page.goto("https://playwright.dev/python")

    docs_link = page.get_by_role("link", name="Docs")

    expect(docs_link).to_have_class("navbar__item navbar__link") # exact attribute value matching
    expect(docs_link).to_have_class(re.compile(r"navbar__item")) # partial attribute matching

    expect(docs_link).not_to_have_class("navbar__item navbar__link")  # exact attribute value matching
    expect(docs_link).not_to_have_class(re.compile(r"^navbar__item"))  # partial attribute matching

    expect(docs_link).to_have_attribute("href", "attribute-value") # attribute-value is optional
    expect(docs_link).not_to_have_attribute("href", "attribute-value") # attribute-value is optional
```

# Input Field assertions

```python

from playwright.sync_api import Page, expect

DOCS_URL=("https://playwright.dev/python/docs/intro")

def test_page_has_get_started_link(page: Page):
    page.goto("https://playwright.dev/python")

    input = page.get_by_placeholder("Search docs")
    # Note that the input is hidden before button click

    expect(input).to_be_hidden()

    # search button
    search_btn = page.get_by_role("button", name="Search")
    search_btn.press("Control+KeyK")
    # should pop the search menu

    expect(input).to_be_editable()
    expect(input).to_be_empty()

    # type some query in the input
    query = "assertions"
    input.fill(query)

    # check the input value
    expect(input).to_have_value(query)
```

# Checkbox assertions

```python

from playwright.sync_api import Page, expect

DOCS_URL=("https://playwright.dev/python/docs/intro")

def test_page_has_get_started_link(page: Page):
    page.goto("https://bootswatch.com/default")

    checked_checkbox = page.get_by_label("Checked checkbox")
    default_checkbox = page.get_by_label("Default checkbox")

    expect(checked_checkbox).to_be_checked()
    expect(default_checkbox).not_to_be_checked()
```

# Option menu assertions

```python
from playwright.sync_api import Page, expect


def test_app(page: Page):
    page.goto("https://bootswatch.com/default")
    
    option_menu = page.get_by_label("Example select")
    
    # expect selected option
    expect(option_menu).to_have_value("1")

    multi_option_menu = page.get_by_label("Example multiple select

    # expect selected options - empty list since no option is selected
    expect(multi_option_menu).to_have_values([])
                                          
    multi_option_menu.select_option(["2", "4"])

    # expect selected options
    expect(multi_option_menu).to_have_values(["2", "4"])
```
