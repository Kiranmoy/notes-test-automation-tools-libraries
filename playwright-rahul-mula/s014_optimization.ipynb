# Intercept Resource Requests

* We'll learn some techniques that can be used to optimize the tests. 
* Specifically, we'll learn how to skip over resources our tests don't require.
* That is, if we initiate any page navigation, Playwright goes ahead and loads every resource such as images, fonts, style sheets, etc.

**Example**:

```python

import pytest
from playwright.sync_api import Page, expect, Route

NOT_ALLOWED_RESOURCES = (
    "image",
    "fonts",
    "stylesheets",
    "media",
    "script"
)
def on_route(route: Route):
    if route.request.resource_type in NOT_ALLOWED_RESOURCES:
        route.abort()
    else:
        route.continue_()


@pytest.fixture(autouse=True)
def skip_resources(page: Page):
    page.route(
        url="**",
        handler=on_route
    )


# In this test, we're not bothered about images, style sheets, and any other things.
# We only need the HTML data such as link, buttons, tables, etc.
# So we can set up an ROUTE HANDLER to skip over those resources
def test_page_has_docs_link(page: Page):
    page.goto("https://playwright.dev/python")
    link = page.get_by_role("link", name="docs")
    expect(link).to_be_visible()


# In this test, we're not bothered about images, style sheets, and any other things.
# We only need the HTML data such as link, buttons, tables, etc.
# So we can set up an ROUTE HANDLER to skip over those resources
def test_get_started_link(page: Page):
    page.goto("https://playwright.dev/python")
    link = page.get_by_role("link", name="GET STARTED")
    link.click()
    expect(page).to_have_url("https://playwright.dev/python/docs/intro")

```

# Disabling Javascript

* We'll learn how to disable JavaScript execution in our web pages.
* Before disabling JavaScript, what we can do is skip over the JavaScript files from being loaded by adding `script` to `NOT_ALLOWED_RESOURCES`.
* Now, if we want to disable JavaScript altogether in our web page, then we need to pass a special argument while creating the browser context.

**Example**:

```python
import pytest
from playwright.sync_api import Page, expect, Route

NOT_ALLOWED_RESOURCES = (
    "image",
    "fonts",
    "stylesheets",
    "media",
    "script"
)

def on_route(route: Route):
    if route.request.resource_type in NOT_ALLOWED_RESOURCES:
        route.abort()
    else:
        route.continue_()

@pytest.fixture
def browser_context_args():
    return {"java_script_enabled": False}

@pytest.fixture(autouse=True)
def skip_resources(page: Page):
    page.route(
        url="**",
        handler=on_route
    )


# In this test, we're not bothered about images, style sheets, and any other things.
# We only need the HTML data such as link, buttons, tables, etc.
# So we can set up an ROUTE HANDLER to skip over those resources
def test_page_has_docs_link(page: Page):
    page.goto("https://playwright.dev/python")
    link = page.get_by_role("link", name="docs")
    expect(link).to_be_visible()


# In this test, we're not bothered about images, style sheets, and any other things.
# We only need the HTML data such as link, buttons, tables, etc.
# So we can set up an ROUTE HANDLER to skip over those resources
def test_get_started_link(page: Page):
    page.goto("https://playwright.dev/python")
    link = page.get_by_role("link", name="GET STARTED")
    link.click()
    expect(page).to_have_url("https://playwright.dev/python/docs/intro")
```

# Parallelism - Run Tests in Parallel

* By default, pytest runs tests sequentially.
* To run multiple tests in parallel, we need to install a plugin: `pip install pytest-xdist`.
* `pytest --numprocesses auto`: This will decide the number of processes to run in parallel by considering the number of threads in our CPU.
* `pytest -n 3`: Run 3 processes in parallel, that is, one process per thread.
