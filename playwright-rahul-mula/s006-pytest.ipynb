# Pytest Fixture

* Fixture is a simple function that allows data/state to be shared among test functions.
* Fixture cannot be called directly as a function to get the data. It will throw an error if attempted.
* To get the data in the test functions, we need to pass the fixture function name as an argument to the test functions.

```python
import json
import report
import pytest

# By default fixture scope=function/method
@pytest.fixture
def report_json():
    print("\n[ Fixture ]: requested...")
    report.generate_report()

    with open("report.json") as file:
        print("[ Fixture ]: ...return report data")
        return json.load(file)

# passing fixture name as an argument
def test_report_json(report_json):
    print("[ Test ]: recieved -", report_json)
    assert type(report_json) == dict

# passing fixture name as an argument
def test_report_fields(report_json):
    print("[ Test ]: recieved -", report_json)
    assert "timestamp" in report_json
    assert "status" in report_json
```

# Fixture Scope

In pytest, fixture scope determines how often a fixture is created and destroyed during a test session. 
* The default scope is `function`, meaning it's created and destroyed for each test function.
* Other scopes include `class`, `module`, `package`, and `session`, which control the fixture's lifetime at the **class**, **module**, **package**, or **session** level, respectively. 


**`scope=function`  (default)**:
* The fixture is created and destroyed at the beginning and end of each test function.
* This is the most common scope and is suitable for fixtures that need to be reset for each test.

**`scope=class`**:
* The fixture is created once for each test class and destroyed after the last test in the class is executed.
* This is useful when you need a fixture to persist across multiple tests within a class, but not across different classes.

**`scope=module`**:
* The fixture is created once for the entire module (test file) and destroyed after the last test in that module is executed.
* This is useful for fixtures that require significant setup or teardown, and you want to share them across multiple tests within a module.

**`scope=package`**:
* The fixture is created once for all tests within a package (a collection of modules) and destroyed at the end of the test package.

**`scope=session`**:
* The fixture is created only once for the entire test session and is destroyed at the end of the session.
* This is suitable for fixtures that need to persist across multiple test runs and are not tied to any specific module or package. 


```python

# Fixture with default scope (function)
@pytest.fixture
def my_fixture():
    
    # Some setup code
    print("Setting up my_fixture (function scope)")
    value = "Initial value"
    
    # Some teardown code
    yield value
    print("Tearing down my_fixture (function scope)")

# Fixture with module scope
@pytest.fixture(scope="module")
def module_fixture():
    # Some setup code
    print("Setting up module_fixture (module scope)")
    value = "Module value"
    
    # Some teardown code
    yield value
    print("Tearing down module_fixture (module scope)")
```
