# Locust Master-Slave architecture model



---



---

**Command Master Node**: `locust -f <filename> --master [ --expect-workers=n -headless ]`

**Command Slave/Worker Node**: `locust -f <filename> --worker --master-host <ip> --master-port <port>`

# Locust execution with config file

## Single node execution model

**`basic_config.yml`**

```yaml
locustfile: locust_test_load.py
host:  http://automationpractice.com
users: 10
spawn-rate: 5
headless: true
run-time: 15s
logfile: run.log
```

**Command**: `locust --config=basic_config.yml`

## Master Slave execution model

**`master_config.yml`**

```yaml
locustfile: locust_test_load.py
host:  http://automationpractice.com
master: true
expect-workers: 2
users: 20
spawn-rate: 5
headless: true
run-time: 30s
logfile: run.log
```

**`worker_config.yml`**

```yaml
locustfile: locust_test_load.py
worker: true
master-host: 127.0.0.1
master-port: 5557
```



**Command**: 

```bash
locust --config=master_config.yml # execute on master node

locust --config=worker_config.yml # execute on worker node
locust --config=worker_config.yml # execute on worker node
```
